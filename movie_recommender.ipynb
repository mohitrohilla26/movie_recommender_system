```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv("tmdb_5000_movies.csv")
credits=pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies=movies.merge(credits)
```

```python
movies.head()
```

```python
# drop unneccessary elements
movies= movies[['id','title', 'genres','keywords','overview','production_companies','cast','crew']]
```

```python
movies.head()
```

```python
movies.info()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.shape
```

```python
movies.duplicated().sum()
```

```python
import ast
```

```python
def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert_production(obj):
  L=[]
  counter =0
  for i in ast.literal_eval(obj):
    if counter != 1:
      L.append(i['name'])
      counter +=1
    else:
      break
  return L
```

```python
movies['production_companies']=movies['production_companies'].apply(convert_production)
```

```python
movies.head()
```

```python
def convert_cast(obj):
  L=[]
  counter =0
  for i in ast.literal_eval(obj):
    if counter != 3:
      L.append(i['name'])
      counter +=1
    else:
      break
  return L
```

```python
movies['cast']=movies['cast'].apply(convert_cast)
```

```python
movies.head()
```

```python
def convert_crew(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew']=movies['crew'].apply(convert_crew)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['production_companies']=movies['production_companies'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['genres'] + movies['keywords'] + movies['overview'] + movies['production_companies'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_movies=movies[['id','title','tags']]
```

```python
new_movies.head()
```

```python
new_movies['tags']=new_movies['tags'].apply(lambda x: " ".join(x))
```

```python
new_movies.head()
```

```python
new_movies['tags']=new_movies['tags'].apply(lambda x: x.lower())
```

```python
new_movies.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]

  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_movies['tags']=new_movies['tags'].apply(stem)
```

```python
#vectozise tags
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors=cv.fit_transform(new_movies['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
  movie_index=new_movies[new_movies['title']==movie].index[0]
  distances=similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)), reverse = True, key= lambda x: x[1])[1:6]

  for i in movies_list:
    print(new_movies.iloc[i[0]].title)
```

```python
recommend('The Notebook')
```
